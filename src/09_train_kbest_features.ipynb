```python
# libraries
from pandas import read_csv
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import plot_roc_curve

from sklearn import tree

from sklearn.naive_bayes import GaussianNB

from sklearn import svm

from sklearn.neighbors import KNeighborsClassifier
```

```python
selected_cols_kbest_df = read_csv('../dataset/df_features_kbest.csv')
selected_cols_kbest_df.index = selected_cols_kbest_df.ID
```

# TRAIN / TEST SPLIT

```python
x = selected_cols_kbest_df.drop(['ID', 'TARGET'], axis=1)
y = selected_cols_kbest_df.TARGET

# Definindo o tamanho do conjunto de dados
test_size = 0.3

# Dividindo os dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = test_size, random_state = 123)
```

```python
x_train.shape
```

```python
x_test.shape
```

# TRAIN

### Logistic Regression with SELECTKBEST selected features

```python
# Criando o modelo
model = LogisticRegression()
model.fit(x_train, y_train)
predict = model.predict(x_test)
```

```python
#extracting true_positives, false_positives, true_negatives, false_negatives
tn, fp, fn, tp = confusion_matrix(y_test, predict).ravel()
print("True Negatives: ",tn)
print("False Positives: ",fp)
print("False Negatives: ",fn)
print("True Positives: ",tp)
print()

#Accuracy
Accuracy = (tn+tp)*100/(tp+tn+fp+fn) 
print("Accuracy {:0.2f}%:", format(Accuracy))

plot_confusion_matrix(model, x_test, y_test, normalize= 'true')
plot_roc_curve(model, x_test, y_test)
```
