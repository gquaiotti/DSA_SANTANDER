```python
# libraries
from pandas import read_csv
from pandas import concat
from pandas import DataFrame
from sklearn.preprocessing import MinMaxScaler
```

## SCALE

```python
df = read_csv('../dataset/df_balance.csv')
df.index = df.ID
```

```python
scaler = MinMaxScaler( feature_range = (0, 1) )

scaled_df = DataFrame(scaler.fit_transform( df.drop( ['ID', 'TARGET'], axis=1 ) ), index = df.drop( ['ID', 'TARGET'], axis=1 ).index, columns = df.drop( ['ID', 'TARGET'], axis=1 ).columns)
scaled_df = concat( [df.ID, scaled_df], axis=1 )
scaled_df = concat([scaled_df, df.TARGET], axis=1)
df = scaled_df
df
```

```python
df.to_csv('../dataset/df_scale.csv', index=False)
```
